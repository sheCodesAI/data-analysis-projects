```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
dsa_marks = [90, 92, 96, 98]
ml_marks = [93, 95, 98, 100]

plt.plot(x, dsa_marks, label='dsa marks', color='m', linewidth=2,
         marker='*', linestyle='--', markersize=20)

plt.plot(x, ml_marks, label='ml marks')

plt.title("Marks corresponding to no of hrs studied",
          fontdict={'fontname': 'comic sans ms', 'fontweight': 20, 'color': 'blue'})

plt.xlabel("No of hrs studied",
           fontdict={'fontname': 'times new roman', 'fontweight': 5, 'color': 'red'})

plt.ylabel("Marks",
           fontdict={'fontname': 'times new roman', 'fontweight': 5, 'color': 'red'})

plt.xticks([1,2,3,4])
plt.yticks([90, 92, 94, 96, 98, 100])

plt.legend(loc="lower right")

plt.savefig("marks.png")

plt.show()
```
