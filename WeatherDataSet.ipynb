````python
# 9) find all instances when 'Snow' was recorded

# three methods to ans :
# 1) value_counts()
# 2) filtering
# 3) str.contains
````

````python
# value_counts()
data.WeatherCondition.value_counts()
````

````python
# filtering
data[data.WeatherCondition == 'Snow']
````

````python
# str_contains
data[data['WeatherCondition'].str.contains('Snow')].tail(10)

# we use str.contains bcz many 2 letter word also there, there snow word is also included , so only to include 'Snow' we put '.str.contains('Snow')'
````

````python
# 10) find all instances when 'Wind speed is above 24' and 'Visibility is 25'
````

````python
data[(data['Wind Speed_km/h'] > 24) & (data.Visibility_km == 25)]
````

````python
# 11) what is the Mean Value of each column against each 'WeatherCondition' ?
````

````python
# 14) Find all the instances when 'Weather is Clear' or 'Visibility is above 40'
````

````python
# 15) Find all instances when :

# A. 'Weather is Clear' and 'Relative Humidity is > 50'
# OR
# B. 'Visibility is above 40'
````

````python
data[
    ((data['WeatherCondition'] == 'Clear') & (data['Rel Hum_%'] > 50)) |
    (data['Visibility_km'] > 40)
]
````
